## Table Analysis

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

### Bronze

**customers_bronze**

```python
%sql
select * from project.dlt_bank.customers_bronze
```

```python

cols_to_check = [ "gender","city","status","preferred_channel","occupation", "income_range","risk_segment"]
df = spark.read.table("project.dlt_bank.customers_bronze")
for c in cols_to_check:
    print(f"Distinct values and counts for column: {c}")
    (
        df.groupBy(col(c))
          .agg(count("*").alias("value_count"))
          .orderBy("value_count", ascending=False)
          .display()
    )
```

**accounts_bronze**
